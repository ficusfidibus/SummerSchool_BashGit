#### 2.8 Permissions

![Figure 2 - Permission explanations](https://www.digitalocean.com/api/static-content/v1/images?src=https%3A%2F%2Fassets.digitalocean.com%2Farticles%2Flinux_basics%2Fmode.png&token=11cf955620aa854e140ef39a8fe4056cb8ad4c95e06e23949bbd449e098c20ad)

Source:
https://www.digitalocean.com/community/tutorials/an-introduction-to-linux-permissions

Reading the string from `ls -l` left to right: the first character is the
file type (`-` file, `d` directory), then three groups of `rwx` for
**owner**, **group**, and **others**.

- `-rw-------`: only the owner can read/write it
- `-rwxr-xr-x`: executable by every user on the system ("world-executable")
- `-rw-rw-rw-`: modifiable by every user on the system ("world-writable")
- `drwxr-xr-x`: a directory every user can read and access
- `drwxrwx---`: a directory modifiable (including contents) by owner and group
- `drwxr-x---`: a directory accessible by its group

To **change** permissions, use `chmod`:

- `chmod a-w <file>` — remove write permission for everyone (`a`)
- `chmod u+w <file>` — add write permission back for the owner (`u`)
- `chmod 444 <file>` — numeric form: read-only for owner/group/others

````python
%%bash

cd ~/bash_practice
ls -l scratch_file.txt
chmod a-w scratch_file.txt
ls -l scratch_file.txt
chmod u+w scratch_file.txt
ls -l scratch_file.txt
````

#### 2.9 Pipes & redirection

You can combine commands using the pipe symbol `|`, feeding the output of
one command as input to the next. `>` writes/overwrites a file with the
output, `>>` appends to it.

Some commands can also act *directly* on each result they find (`-exec`),
or hand results to another command as arguments (`xargs`):

- `find <location> ... -exec <command> {} +` — runs `<command>` on the
  files `find` matched. `{}` is a placeholder that gets replaced by the
  matched path(s); the trailing `+` batches as many paths as possible into
  a single call to `<command>` (use `\;` instead of `+` to run the command
  once *per file* — slower, but occasionally necessary). E.g.
  `find . -name "*.txt" -exec realpath {} +` prints the **absolute path**
  of every `.txt` file found.
- `<command> | xargs <other_command>` — takes whatever `<command>` printed
  (one item per line) and feeds it to `<other_command>` as **arguments**,
  rather than as piped-in text. This matters because some commands (like
  `grep`) expect filenames as arguments, not text on standard input —
  `find . -name "*.txt" | xargs grep -l "failed"` runs `grep -l "failed"`
  directly on the list of `.txt` files that `find` located.

````python
%%bash

cd ~/bash_practice/project_alpha
find . -name "*.txt" | wc -l # number of files ending with .txt
find . -name "*.txt" | grep "sample" | wc -l # number of .txt files with "sample" in the name
find . -name "*.txt" -exec realpath {} + >> realpaths_to_file.txt # write out absolute paths, one per line
cat raw_data/notes.txt | head -2
cat raw_data/notes.txt | wc -l
````

#### 2.10 Shell scripts

A **shell script** is just a text file full of Bash commands, saved so you
can run the whole sequence as one program instead of retyping every line.
This is exactly what the capstone project below asks you to build.

- `#!/bin/bash` — the **shebang**, always the very first line of a script;
  it tells the system which interpreter should execute the file.
- `chmod +x <script>` — makes the file executable (see **2.8** above).
- `./<script>` — runs it. The `./` means "look in the current directory";
  without it, Bash searches your `$PATH` and likely won't find the script.
- `variable=value` (no spaces around `=`!), then `$variable` or
  `${variable}` — store a value and reuse it later in the script.
- `` $(command) `` — **command substitution**: runs `command` and drops its
  output in place, e.g. `count=$(find . -name "*.txt" | wc -l)` stores the
  number of `.txt` files found in the variable `count`.

**Creating the script file**

- In a real terminal you'd normally open a text editor, e.g. `nano script.sh`,
  type the commands, then save and exit.
- `nano` doesn't work inside a notebook cell (it's interactive), so here we
  write the whole file in one go with a **heredoc**:
  ```bash
  cat > script.sh << 'EOF'
  #!/bin/bash
  echo "Hello, world!"
  EOF
  ```
  Everything between `<< 'EOF'` and the closing `EOF` is written to
  `script.sh` exactly as typed.
- Since Google Colab runs on a real Ubuntu VM, you can also create or edit
  the file **by hand**: click the folder icon in the left sidebar to open
  the file browser, navigate to the right folder, right-click → *New
  file*, then double-click it to open Colab's built-in text editor. Just
  make sure you save it to the same path your notebook is using (e.g.
  inside `~/bash_practice`).

````python
%%bash

cd ~/bash_practice
cat > greet.sh << 'EOF'
#!/bin/bash
name="World"
echo "Hello, $name!"
lines=$(wc -l < greet.sh)
echo "This script itself has $lines lines."
EOF
chmod +x greet.sh
./greet.sh
rm greet.sh
````
